```python
cookie = [] #insert request cookies within{}
def getAmazonSearch(search_query):
    url="https://www.amazon.in/s?k="+search_query
    print(url)
    page=requests.get(url,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchasin(asin):
    url="https://www.amazon.in/dp/"+asin
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchreviews(review_link):
    url="https://www.amazon.in"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
product_names=[]
response=getAmazonSearch('smart+watch+for+women')
soup=BeautifulSoup(response.content)
for i in soup.findAll('span',{'class':'a-size-medium a-color-base a-text-normal'}):
    product_names.append(i.text)
```

```python
data_asin=[]
response= getAmazonSearch('smart+watch+for+women')
soup=BeautifulSoup(response.content)
for i in soup.findAll("div",{'class':"s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col s-widget-spacing-small sg-col-12-of-16"}):
    data_asin.append(i['data-asin'])
```

```python
response.status_code
```

```python
data_asin
```

```python
len(data_asin)
```

```python
link=[]
for i in range(len(data_asin)):
    response=Searchasin(data_asin[i])
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])
```

```python
len(link)
```

```python
comment=[]
for j in range(len(link)):
    for k in range(20):
        response=Searchreviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            comment.append(i.text)
```

```python
rev={'comments':comment}
review_data=pd.DataFrame.from_dict(rev)
pd.set_option('max_colwidth',800)
review_data.head(5)
```

```python
import re
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd
```

```python
with open("extract_revivews.csv","r",encoding='utf8')as input:
    Amazon_reviews = input.read()
type(Amazon_reviews)
```

```python
rev_string = Amazon_reviews
rev_string = re.sub("[^A-Za-z" "]+"," ",rev_string).lower()
rev_string = re.sub("[0-9" "]+"," ",rev_string)

reviews_words = rev_string.split(" ")
```

```python
stop_words = stopwords.words('english')

reviews_words = [w for w in reviews_words if not w in stop_words]
```

```python
rev_string = " ". join(reviews_words)
```

```python
rev_string
```

```python
wordcloud_rev = WordCloud(background_color='white',
                         max_words = 100
                         ).generate(rev_string)

plt.imshow(wordcloud_rev)
plt.axis("off")
plt.show()
```
